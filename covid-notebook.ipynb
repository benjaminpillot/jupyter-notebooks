# Download data from John Hopkins University and plot

#### Import covid package and pyplot:

```python
from covid.plotting import plot_cases, get_world_map, plot_pace
from matplotlib import pyplot as plt
import datetime
```

## Display prevalence on a leaflet map 

Display death prevalence among countries:

```python
wmap = get_world_map(case_type="deaths", normalized=True, date=datetime.datetime(2020,4,4))
```

```python
wmap
```

Display prevalence of confirmed cases:

```python
wmap = get_world_map(case_type="confirmed cases", normalized=True, date=datetime.datetime(2020,4,4))
```

```python
wmap
```

#### Resize figure display:

```python
plt.rcParams['figure.figsize'] = [34/2.54, 18/2.54]
```

## Plot cumulated deaths in France against an exponential where deaths double every 3 days

```python
plot_cases(["France"], case_type="deaths", doubling_factor=3, plot_type="log", 
           normalized=True, min_prevalence=0.01)
```

## Compare different country evolutions
#### 1. Plot cumulated deaths in US, Brazil, France and Australia against an exponential where cases double every 3 days:

```python
plot_cases(["US", "Brazil", "France", "Australia"], case_type="deaths", doubling_factor=3, 
           plot_type="log", min_number_of_cases=10)
```

#### 2. Plot daily increase of confirmed cases in 6 different countries:

```python
plot_cases(["Netherlands", "Brazil", "Italy", "Germany", "France", "Spain"], 
           case_type="confirmed cases", plot_type="daily")
```

## Plot prevalence pace 

Let's plot Covid-19 spreading pace in different countries

```python
plot_pace(["China", "Iran", "Italy", "France", "US"], case_type="deaths")
```

We can also plot the prevalence pace and set all the curves to the same origin, so that we may compare how Covid-19 is spreading in each country.

```python
plot_pace(["France", "US", "Italy", "Iran", "Germany", "Spain", "United Kingdom", "Netherlands"], 
          case_type="deaths", do_offset=True, offset_value=0.01, min_prevalence=0.001)
```

```python
plot_pace(["France", "US", "Italy", "Iran", "Germany", "Spain", "United Kingdom", "Netherlands", "Sweden"], 
          case_type="deaths", plot_type='normal', do_offset=True, offset_value=0.01, min_prevalence=0.001)
```

Compare countries from both hemispheres:

```python
plot_pace(["France", "Australia"], 
          case_type="deaths", plot_type='normal', do_offset=True, offset_value=0.001)
```
