## Plot cumulated cases against an exponential where deaths double every _n_ days

### Use of `plot_cases` function
1. Required parameters
    * `countries`: list of valid countries/regions
    * `case_type`: {"death[s]", "confirmed [cases]", "recovered [cases]"}
2. Optional parameters
    * `doubling_factor`: number of days between each case doubling (by default: 2)
    * `plot_type`: {"log", "daily"}
    * `normalized`: if `True`, show prevalence, otherwise show absolute cases
    * `min_prevalence`: minimum prevalence for cases to be plotted (by default all cases are plotted)
    * `min_number_of_cases`: minimum number of cases for cases to be plotted

```python
plot_cases(["France"], case_type="deaths", doubling_factor=3, plot_type="log", 
           normalized=True, min_prevalence=0.01)
```

## Compare different country evolutions
#### 1. Plot cumulated deaths in US, Brazil, France and Australia against an exponential where cases double every 3 days:

```python
plot_cases(["US", "Brazil", "France", "Australia"], case_type="deaths", doubling_factor=3, 
           plot_type="log", min_number_of_cases=10)
```

#### 2. Plot daily increase of confirmed cases in 6 different countries:

```python
plot_cases(["Netherlands", "Brazil", "Italy", "Germany", "France", "Spain"], 
           case_type="confirmed cases", plot_type="daily")
```

## Plot prevalence pace 

### Use of `plot_pace` function
1. Required parameters
    * `countries`: list of countries/regions
    * `case_type`: {"death[s]", "confirmed [cases]", "recovered [cases]"}
2. Optional parameters
    * `normalized`: if `True`, show prevalence, absolute number of cases otherwise
    * `plot_type`: {"log", "normal"}
    * `do_offset`: if `True`, readjust plots so that countries start at the same prevalence origin
    * `offset_value`: if `do_offset == True`, prevalence of number of cases value for which plots must be readjusted
    * `min_number_of_cases`: minimum number of cases for which data must be plotted (by default, all cases are plotted)
    * `min_prevalence`: minimum prevalence for which data must be plotted

#### Let's plot Covid-19 spreading pace in different countries:

```python
plot_pace(["China", "Iran", "Italy", "France", "US"], case_type="deaths")
```

#### We can also plot the prevalence pace and set all the curves to the same origin, so that we may compare how Covid-19 is spreading in each country.

```python
plot_pace(["France", "US", "Italy", "Iran", "Germany", "Spain", "United Kingdom", "Netherlands"], 
          case_type="deaths", do_offset=True, offset_value=0.01, min_prevalence=0.001)
```

```python
plot_pace(["France", "US", "Italy", "Iran", "Germany", "Spain", "United Kingdom", "Netherlands", "Sweden"], 
          case_type="deaths", plot_type='normal', do_offset=True, offset_value=0.01, min_prevalence=0.001)
```

#### Compare countries from both hemispheres:

```python
plot_pace(["France", "Australia"], 
          case_type="deaths", plot_type='normal', do_offset=True, offset_value=0.001)
```

#### Plot pace of confirmed cases for overseas regions 

Use preferentially absolute cases as population is not always available for all regions
