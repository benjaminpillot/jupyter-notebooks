### Download data from John Hopkins University and plot

#### Import covid package and pyplot:

```python
from covid.plotting import plot_cases, get_world_map
from matplotlib import pyplot as plt
import datetime
```

#### Display death prevalence on a leaflet map 

```python
wmap = get_world_map(case_type="deaths", normalized=True)
```

#### Resize figure display:

```python
plt.rcParams['figure.figsize'] = [34/2.54, 18/2.54]
```

#### Plot cumulated deaths in United Kingdom against an exponential where deaths double every 3 days:

```python
plot_cases(["France"], case_type="deaths", doubling_factor=3, plot_type="log", min_number_of_cases=10)
```

#### Compare different country evolutions
1. Plot cumulated deaths in Brazil and US against an exponential where cases double every 3 days:

```python
plot_cases(["US", "Brazil"], case_type="deaths", doubling_factor=3, plot_type="log")
```

2. Plot daily increase of confirmed cases in 6 different countries:

```python
plot_cases(["Netherlands", "Brazil", "Italy", "Germany", "France", "Spain"], case_type="confirmed", plot_type="daily")
```
